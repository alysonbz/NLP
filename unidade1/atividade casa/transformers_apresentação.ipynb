# Hugging Face Transformers
### A biblioteca Hugging Face Transformers é uma poderosa ferramenta de processamento de linguagem natural (NLP), com suporte para diversos idiomas, incluindo o português do Brasil (pt-BR). Criada para ser utilizada com diferentes modelos pré-treinados, a Hugging Face Transformers oferece um conjunto de módulos para realizar tarefas como análise de sentimentos, tradução, geração de texto, reconhecimento de entidades nomeadas (NER), e muito mais. Embora tenha começado com foco no inglês, possui diversos modelos otimizados para o português.



## Principais Funcionalidades do Hugging Face Transformers
### Processamento de Linguagem Natural (NLP):

*   Tokenização: Divide textos em palavras ou sentenças utilizando tokenizadores compatíveis com vários modelos.
*  Análise de Sentimentos: Detecta o sentimento de uma frase, classificando como positivo, negativo ou neutro (ex.: modelos como BERTimbau para português).
* Reconhecimento de Entidades Nomeadas (NER): Identifica nomes de pessoas, locais, organizações e outras entidades em um texto (ex.: modelos como BERT e XLM-R para português).


### Mineração de Textos:
* POS Tagging: Identifica a estrutura gramatical de uma frase, categorizando palavras como substantivos, verbos, adjetivos e suas funções dentro da sentença.
* Geração de Texto: Utiliza modelos como GPT-2 e T5 para gerar texto automaticamente a partir de prompts.

```python
import spacy
import pandas as pd
from spacy.lang.pt.examples import sentences
from nltk.stem import RSLPStemmer
import nltk
```

```python
nlp = spacy.load("pt_core_news_sm")
doc = nlp(sentences[0])
print(doc.text)
for token in doc:
  print(token.text, token.pos_)
```

```python
dados = [(token.text, token.pos_, token.pos_) for token in doc]

df = pd.DataFrame(dados, columns=["Palavra", "POS", "Significado da Sigla POS"])

pos_significado = {
    "PROPN": "Proper Noun (Nome próprio)",
    "AUX": "Auxiliary (Verbo auxiliar)",
    "VERB": "Verb (Verbo)",
    "DET": "Determiner (Determinante)",
    "NOUN": "Noun (Substantivo)",
    "ADP": "Adposition (Preposição)",
    "NUM": "Numeral (Número)"
}

df["Significado da Sigla POS"] = df["POS"].map(pos_significado)

df
```

### 3° Faça uma implementação para que sua aplicação receba um texto com erros gramaticais em português.
1. Realize a correção do texto
2. Tokenize o texto
3. Remova as stop words
4. Faça uma tabela mostrando 3 colunas para todos os tokens: token , stemming e lematização.

```python
!pip install language-tool-python
!pip install enelvo
```

```python
from enelvo.normaliser import Normaliser

norm = Normaliser(tokenizer='readable', capitalize_inis=True,
                  capitalize_pns=True, capitalize_acs=True,
                  sanitize=False)

texto = "Eu gosto de correr,ela gosta ir pra acadêmia"
resposta = norm.normalise(texto)
print(resposta)
```

```python
doc = nlp(resposta)

import nltk
nltk.download('rslp')

stemmer = RSLPStemmer()

data = []

for token in doc:
    if not token.is_stop and token.is_alpha:
        lemma = token.lemma_
        stem = stemmer.stem(token.text)
        data.append([token.text, stem, lemma])

df = pd.DataFrame(data, columns=["Token", "Stemming", "Lematização"])

print("\nTabela de Tokens, Stemming e Lematização:")
print(df)
```
