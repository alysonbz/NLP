* Contagem de ocorrêcnias

```python
import re
```

```python
string = '''a
         Hello python no Python é mais fácil que outras linguagens que não sejam PYTHON,
         PYthon é mais lega.
         A logo do PytHOn são duas cobras descobri isso agora.
         Te amo Phton
         '''
def qnt_ocorrencias(RE, string):
    lista = re.findall(RE, string, re.I)
    qtd = len(lista)
    print(qtd)

qnt_ocorrencias(RE = r'python', string = string)
```

* Validação de e-mail

1. O nome do usuário (antes do @)
* Pode conter letras maiúsculas e minúsculas (A-Z, a-z).
* Pode conter dígitos (0-9).
* Pode conter pontos (.), sublinhados (_), e hífens (-), mas não consecutivamente ou no início/fim.
    
2. O domínio (após o @):
* Deve começar e terminar com uma letra ou dígito.
* Pode conter letras maiúsculas e minúsculas (A-Z, a-z).
* Pode conter dígitos (0-9).
* Pode conter hífens (-), mas não consecutivamente ou no início/fim.
* Pode ter subdomínios (partes separadas por pontos).

3. A TLD (Top-Level Domain, como .com, .net):
* Deve ser composta por pelo menos duas letras.
* Pode conter apenas letras (A-Z, a-z).


```python


RE = r'([A-za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})'

def email_valido(email):
    if re.fullmatch(RE, email):
        print("E-mail válido")
    else:
        print("E-mail não válido")

email_valido('brunabarretomq@gmail.com')
email_valido('meuemailola@gma**il.com')

```

* Extração de números de telefone

1. Pode conter dígitos (0-9).
2. Pode conter espaços, hífens (-), parênteses e pontos (.) como separadores opcionais.
3. Pode começar com um sinal de mais (+) para códigos de país.
4. Pode ter parênteses para agrupar o código de área.
5. Deve ter um comprimento mínimo e máximo razoável para evitar validação excessivamente permissiva.

```python
texto = '''
12345
Agora você está ligando
3 da manhã, minha cabeça está girando
Puxo o lençol da minha cama
Porque o telefone continua tocando
Tudo que continuo pensando, é "será que eu enlouqueci? "
Mas vou atender pela última vez
número : 85992408798
Não ligue pra mim +55 (11) 91234-5678
call to +55(85)992378890
'''

phone_re_br = r'\+?55?\s?\(?\d{2}\)?[\s-]?\d{4,5}[\s-]?\d{4}'

def telefone_valido(texto):
    numeros = re.findall(phone_re_br, texto)
    print(numeros)

telefone_valido(texto)
```

* Substiuição de palavras

```python
def substituir_palavra(texto, palavra_antiga, palavra_nova):
    padrao = rf'\b{re.escape(palavra_antiga)}\b'
    texto_modificado = re.sub(padrao, palavra_nova, texto)
    return texto_modificado


texto = '''
É verdade que no verão os cachorro
tomam banho nas caixas de água e 
a gente fica contente
Cachorro é cachorro 
Cachorro não é gato 
'''

texto_modificado = substituir_palavra(texto, 'cachorro', 'gato')
print(texto_modificado)
```
