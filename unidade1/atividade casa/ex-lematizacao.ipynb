**Diferença entre Stemming e Lematização**

Stemming e lematização são técnicas utilizadas em processamento de linguagem natural (NLP) para reduzir palavras a suas formas base ou raízes, mas elas operam de maneiras diferentes e têm finalidades distintas.

**Stemming**

O stemming é um processo que corta os sufixos das palavras para reduzir variantes flexionais a uma raiz comum. Esta técnica é baseada em regras heurísticas simples, sem levar em conta o contexto ou o significado da palavra. O objetivo é transformar diferentes formas de uma palavra em uma base comum para fins de análise textual.

Exemplo:

"amando", "amará", "amado" -> "am"
"escolas", "escolar", "escolinha" -> "escol"
O stemming pode resultar em raízes que não são palavras válidas no idioma, como "am" para as diferentes formas da palavra "amar".

Vantagens do Stemming:

Simplicidade e velocidade de execução.
Reduz variações de palavras a uma forma comum, o que pode ser útil para algoritmos que se beneficiam de uma redução na diversidade de palavras.

Desvantagens do Stemming:

Pode produzir resultados que não são palavras válidas.
Não leva em conta o contexto ou o significado das palavras.
Pode ser menos preciso do que a lematização.

-------------------------------------------------------------------------------------

**Lematização**


A lematização é um processo mais sofisticado que considera o contexto e a análise morfológica para reduzir uma palavra à sua forma base ou "lema". Diferente do stemming, a lematização utiliza um dicionário para garantir que a forma reduzida é uma palavra válida no idioma.

Exemplo:

"amando", "amará", "amado" -> "amar"
"escolas", "escolar", "escolinha" -> "escola"
A lematização produz formas base que são palavras reais e têm significado, considerando o contexto em que são usadas.

Vantagens da Lematização:

Produz palavras válidas e significativas.
Leva em conta o contexto e o significado das palavras.
É mais precisa e útil para aplicações que necessitam de compreensão semântica.
Desvantagens da Lematização:

Pode ser mais lenta e complexa devido à necessidade de análise morfológica e consulta a dicionários.
Requer modelos e recursos linguísticos mais completos.

```python
import spacy
from spacy.lang.pt.stop_words import STOP_WORDS
from string import punctuation
import pandas as pd
import nltk
from nltk.stem import RSLPStemmer

# Carregar o modelo spaCy para português
nlp = spacy.load('pt_core_news_sm')

# Download the required resources for RSLPStemmer
nltk.download('rslp')

# Inicializar o stemmer do NLTK
stemmer = RSLPStemmer()

# Função para corrigir o texto (simulação)
def corrigir_texto(texto):
    # Esta função pode ser substituída por uma ferramenta de correção gramatical automática
    correcoes = {
        "vai": "fui",
        "menino": "meninos",
        "estava": "estavam"
    }
    palavras = texto.split()
    texto_corrigido = " ".join([correcoes.get(palavra, palavra) for palavra in palavras])
    return texto_corrigido

# Texto com erros gramaticais
texto_com_erros = "Eu vai para a escola ontem. Os menino estava brincando."

# Correção gramatical
texto_corrigido = corrigir_texto(texto_com_erros)

# Processar o texto corrigido com spaCy
doc = nlp(texto_corrigido)

# Tokenização
tokens = [token.text for token in doc]

# Remoção de stop words e pontuação
tokens_sem_stopwords = [token for token in tokens if token.lower() not in STOP_WORDS and token not in punctuation]

# Stemming e Lematização
tokens_stemming = [stemmer.stem(token) for token in tokens_sem_stopwords]
tokens_lematizados = [token.lemma_ for token in doc if token.text in tokens_sem_stopwords]

# Criar a tabela
data = {
    'Token': tokens_sem_stopwords,
    'Stemming': tokens_stemming,
    'Lematização': tokens_lematizados
}
df = pd.DataFrame(data)

# Exibir resultados
print("Texto com Erros:", texto_com_erros)
print("Texto Corrigido:", texto_corrigido)
print("\nTabela de Tokens, Stemming e Lematização:")
print(df)
```
